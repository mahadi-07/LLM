## OpenAI

```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
GLM_API_KEY=os.getenv('ZAI_API_KEY')
```

```python
client = OpenAI(
    api_key=GLM_API_KEY,
    base_url="https://api.z.ai/api/coding/paas/v4"
)

response = client.chat.completions.create(
    model="glm-5.1",
    messages=[
        {"role": "user", "content": "What is the capital of Bangladesh?"}
    ],
    stream=False
)
```

```python
print(response)
```

```python
import json

print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
```

## Testing MCP

```python
"""
Use LiteLLM Proxy MCP Gateway to call MCP tools.

When using LiteLLM Proxy, you can use the same MCP tools across all your LLM providers.
"""
import openai

client = openai.OpenAI(
    api_key="sk-my-super-secret-key", # paste your litellm proxy api key here
    base_url="http://localhost:4000" # paste your litellm proxy base url here
)
print("Making API request to Responses API with MCP tools")

response = client.responses.create(
    model="glm-5.1",
    input=[
        {
            "role": "user",
            "content": "what is 1 + 1?",
            "type": "message"
        }
    ],
    tools=[
        {
            "type": "mcp",
            "server_label": "litellm",
            "server_url": "litellm_proxy",
            "require_approval": "never"
        }
    ],
    stream=False,
    tool_choice="auto"
)

```

```python
import json

if isinstance(response, openai.Stream):
    for event in response:
        if hasattr(event, 'type') and event.type == "response.output_text.delta":
            print(event.delta, end="", flush=True)
        elif hasattr(event, 'type') and event.type == "response.completed":
            print()
        else:
            print(event)
else:
    # Non-streaming Response object
    for item in response.output:
        if hasattr(item, 'content'):
            for content in item.content:
                if hasattr(content, 'text'):
                    print(content.text)
    print()
    print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
```
